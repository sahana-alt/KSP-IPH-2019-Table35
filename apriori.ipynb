```python
x=dataset.iloc[:10000,:33]
x
```

```python
x.shape
```

```python
records=[]
for i in range(0,10000):
    records.append([str(x.values[i,j])for j in range(0,33)])
records
```

```python
association_rules1=apriori(records,min_support=0.2, min_confidence=0.02, min_lift=1.2,min_length=11)
association_list=list(association_rules1)
association_list
```
